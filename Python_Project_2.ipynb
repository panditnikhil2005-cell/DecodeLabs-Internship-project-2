## **Algorithm / Working**

1. Start the program.
2. Create an empty list to store expenses.
3. Show a menu with options:

   * Add Expense
   * View Expenses
   * Show Total Spent
   * Show Average Expense
   * Delete Expense
   * Exit
4. If the user selects **Add Expense**, take the expense amount as input and store it in the list.
5. If the user selects **View Expenses**, display all expenses with numbering.
6. If the user selects **Show Total Spent**, calculate and display the sum of all expenses.
7. If the user selects **Show Average Expense**, calculate and display the average of all expenses.
8. If the user selects **Delete Expense**, remove the selected expense from the list.
9. If the user selects **Exit**, stop the program.

---

# **PYTHON CODE**

```python
# =========================
# EXPENSE TRACKER APPLICATION
# =========================

expenses = []

while True:
    print("\n========== EXPENSE TRACKER ==========")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Show Total Spent")
    print("4. Show Average Expense")
    print("5. Delete Expense")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        try:
            amount = float(input("Enter expense amount: "))
            expenses.append(amount)
            print("Expense added successfully!")
        except ValueError:
            print("Please enter a valid amount.")

    elif choice == "2":
        if not expenses:
            print("No expenses recorded.")
        else:
            print("\nYour Expenses:")
            for i, amount in enumerate(expenses, start=1):
                print(f"{i}. ₹{amount:.2f}")

    elif choice == "3":
        total = sum(expenses)
        print(f"Total Spent: ₹{total:.2f}")

    elif choice == "4":
        if not expenses:
            print("No expenses available to calculate average.")
        else:
            average = sum(expenses) / len(expenses)
            print(f"Average Expense: ₹{average:.2f}")

    elif choice == "5":
        if not expenses:
            print("No expenses available to delete.")
        else:
            print("\nYour Expenses:")
            for i, amount in enumerate(expenses, start=1):
                print(f"{i}. ₹{amount:.2f}")
            try:
                exp_num = int(input("Enter expense number to delete: "))
                if 1 <= exp_num <= len(expenses):
                    removed = expenses.pop(exp_num - 1)
                    print(f"Expense ₹{removed:.2f} deleted successfully!")
                else:
                    print("Invalid expense number.")
            except ValueError:
                print("Please enter a valid number.")

    elif choice == "6":
        print("Thank you for using Expense Tracker!")
        break

    else:
        print("Invalid choice! Please try again.")
```
